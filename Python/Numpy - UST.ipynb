```python
import time
import numpy as np
arraysize = 1000000
def python_version():
    t1 = time.time()
    x=range(arraysize)
    y=range(arraysize)
    z=[x[i] + y[i] for i in range(len(x))]
    return time.time()-t1

def numpy_version():
    t1 = time.time()
    x=np.arange(arraysize)
    y=np.arange(arraysize)
    z=x+y
    return time.time()-t1
t1=python_version()
t2=numpy_version()

print('numpy in this example is : ' + str(t1/t2) + 'faster')
```
