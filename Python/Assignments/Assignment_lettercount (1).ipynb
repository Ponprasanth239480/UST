```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import re
```

```python
text = open("constitution.txt", "r")
d = dict()
pattern = r'[0-9]'
for line in text:
    line=line.strip()
    words=line.lower()
    new_text = re.sub(r"[^a-zA-Z0-9 ]", "", words)
    new_text1 = re.sub(r"\s+", "", new_text)
    new_text2 = re.sub(pattern, '', new_text1)
    for i in new_text2:
        #print(i)
        for word in i:
            if word in d:
                d[word] = d[word] + 1
            else:
                d[word] = 1
print(d)
letters = list(d.keys())
letter_counts = list(d.values())
print(letters)
print(letter_counts)
```

```python
fig = plt.figure(figsize=(14,8))

plt.bar(range(len(d)), letter_counts, tick_label=letters)
plt.title('letters vs letter_counts bar plot')
plt.xlabel('letters')
plt.ylabel('letter_counts')
plt.show()
```
